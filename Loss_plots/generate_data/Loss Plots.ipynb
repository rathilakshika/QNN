```python
import matplotlib.pyplot as plt
import numpy as np

loss1Layer = np.zeros((100,100))
for i in range(100):
    file = 'Layer1Easy/quantum_loss_easy_layer1h_%d.npy'%i
    loss1Layer[i] += np.load(file, allow_pickle=True)
    
loss2Layers = np.zeros((100,100))
for i in range(100):
    file = 'Layer2Easy/quantum_loss_easy_layer2h_%d.npy'%i
    loss2Layers[i] += np.load(file, allow_pickle=True)
    
loss3Layers = np.zeros((100,100))
for i in range(100):
    file = 'Layer3Easy/quantum_loss_easy_layer3h_%d.npy'%i
    loss3Layers[i] += np.load(file, allow_pickle=True)
    
loss4Layers = np.zeros((100,100))
for i in range(100):
    file = 'Layer4Easy/quantum_loss_easy_layer4_%d.npy'%i
    loss4Layers[i] += np.load(file, allow_pickle=True)
    
loss5Layers = np.zeros((100,100))
for i in range(100):
    file = 'Layer5Easy/quantum_loss_easy_layer5_%d.npy'%i
    loss5Layers[i] += np.load(file, allow_pickle=True)
```

```python
stddevs =[]
averages =[]

sd1 = np.std(loss1Layer, axis=0)
av1 = np.average(loss1Layer, axis=0)
plt.plot(range(100), av1, label='quantum neural network') #, color=groen)
plt.fill_between(range(100), av1+np.array(sd1), av1-np.array(sd1), alpha=0.1)
stddevs.append(sd1)
averages.append(av1)

sd2 = np.std(loss2Layers, axis=0)
av2 = np.average(loss2Layers, axis=0)
plt.plot(range(100), av2, label='quantum neural network') #, color=groen)
plt.fill_between(range(100), av2+np.array(sd2), av2-np.array(sd2), alpha=0.1) #, color=groen)
stddevs.append(sd2)
averages.append(av2)

sd3 = np.std(loss3Layers, axis=0)
av3 = np.average(loss3Layers, axis=0)
plt.plot(range(100), av3, label='quantum neural network') #, color=groen)
plt.fill_between(range(100), av3+np.array(sd3), av3-np.array(sd3), alpha=0.1) #, color=groen)
stddevs.append(sd3)
averages.append(av3)

sd4 = np.std(loss4Layers, axis=0)
av4 = np.average(loss4Layers, axis=0)
plt.plot(range(100), av4, label='quantum neural network') #, color=groen)
plt.fill_between(range(100), av4+np.array(sd4), av4-np.array(sd4), alpha=0.1) #, color=groen)
stddevs.append(sd4)
averages.append(av4)

sd5 = np.std(loss5Layers, axis=0)
av5 = np.average(loss5Layers, axis=0)
plt.plot(range(100), av5, label='quantum neural network') #, color=groen)
plt.fill_between(range(100), av5+np.array(sd5), av5-np.array(sd5), alpha=0.1) #, color=groen)
stddevs.append(sd5)
averages.append(av5)

legendList = ['{}'.format(i) for i in ["1 Layer", "2 Layers","3 Layers","4 Layers","5 Layers"]]
plt.legend(legendList)

plt.xlabel("Number of Iterations")
plt.ylabel("Loss")
plt.title("Training (Data Reuploading, Qubits = 2, Easy Model)")
```
