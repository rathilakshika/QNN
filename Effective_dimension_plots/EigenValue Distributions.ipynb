```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import interpolate

rooi = np.array([255, 29, 0])/255
blou = np.array([0, 150, 236])/255
groen = np.array([0,208,0])/255
```

```python
fhat_1layer = np.load("Data/2qubits_1layer_f_hats_pauli.npy")

e1=[]

for i in range(100):
    e1.append(np.linalg.eigh(fhat_1layer[i])[0])

e1 = np.average(e1, axis=0)

counts, bins = np.histogram(e1, bins=np.linspace(np.min(e1), np.max(e1), 5))
cumulative = counts/np.sum(counts)
plt.bar(bins[:-1], cumulative)
plt.ylabel("Normalised Counts")
plt.xlabel("Eigenvalue size")
plt.title("1 Layer, Easy Model")
```
