Given feature vector : X

```python
from qiskit import *
from math import pi
```

```python
def make_circuit(X):
    qf = QuantumCircuit(len(X), len(X))
    for i in range(len(X)):
        qf.h(i)
    return qf
```

```python
def add_layerRZ(X, qf):
    for i in range(len(X)):
        qf.rz(X[i], i)
    return qf
```

```python
def add_layerRZZ(X, qf):
    for i in range(0, len(X)-1):
        for j in range(i+1, len(X)):
            qf.cx(i, j)
            qf.rz((pi-X[i])*(pi-X[j]), j)
            qf.cx(i, j)
            qf.barrier()
    return qf
```

```python
X = [1,-0.5,-1,0]
```

```python
add_layerRZZ(X, add_layerRZ(X,make_circuit(X))).draw()
```
