```python
from qiskit import *
from math import pi
import numpy as np
from qiskit.aqua.operators import CircuitStateFn
#from qiskit.aqua.operators import *
from qiskit.aqua.operators.gradients import QFI
```

```python
numQubits = 4
```

```python
theta1 = circuit.Parameter('θ1') #X = [1,-0.5,-1,0]
```

```python
theta2 = circuit.Parameter('θ2')
```

```python
theta3 = circuit.Parameter('θ3')
```

```python
def make_circuit_feature():
    qf = QuantumCircuit(QuantumRegister(numQubits)) #, numQubits)
    for i in range(numQubits):
        qf.h(i)
    qf.barrier()
    return qf
```

```python
def add_layerRZ(X, qf):
    for i in range(numQubits):
        qf.rz(theta1, i)
    qf.barrier()
    return qf
```

```python
def add_layerRZZ(X, qf):
    for i in range(0, numQubits-1):
        for j in range(i+1, numQubits):
            qf.cx(i, j)
            qf.rz((pi-theta1)*(pi-theta1), j)
            qf.cx(i, j)
            qf.barrier()
    return qf
```

```python
qf = add_layerRZZ(theta1, add_layerRZ(theta1,make_circuit_feature()))
```

```python
qf.draw('mpl')
```

```python
def add_layerRY(qv, theta):
    qv.ry(theta, range(numQubits))
    qv.barrier()
    return qv
```

```python
def add_cnots(qv):
    for i in range(0, numQubits-1):
        for j in range(i+1, numQubits):
            qv.cx(i, j)
    qv.barrier()
    return qv
```

```python
qv = add_layerRY(add_cnots(add_layerRY(qf, theta2)), theta3)
```

```python
qv.draw('mpl')
```

```python
state = CircuitStateFn(primitive=qv, coeff=1.)
```

```python
params = [theta1, theta2, theta3]
```

```python
qfi = qiskit.aqua.operators.gradients.QFI(qfi_method='lin_comb_full').convert(operator=state, params=params)
```

```python
values_dict = {theta1: pi / 4, theta2: 0.1, theta3: 0.1}     #[0.1, -1, -0.1, 1], theta3: [0.1, -1, -0.1, 1]
```

```python
qfi_result = qfi.assign_parameters(values_dict).eval()
```

```python
print('full QFI \n', np.real(np.array(qfi_result)))
```
